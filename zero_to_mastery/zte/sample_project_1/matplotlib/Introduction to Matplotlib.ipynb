## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot",
      xlabel= "x-axis",
      ylabel= "y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

```python
# Make a plot from a dictionary
nut_butter_prices = { "Almond butter":10 ,
                     "Peanut butter":8 ,
                     "Cashew butter":12 }
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Chris's Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Chris's Nut Butter Store",
       ylabel="Price ($)");
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1 using tuples
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplot option 2 using indexes
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(x);
```

### which one should we use? (pyplot vs matplotlib OO method)

* When plotting somthing quickly, kay to use the pyplot method
* When plotting somthing more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
## OO method  mixed with pyplot

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100]) # xlim shift+tab to see what it does
```

```python
over_50.target.values
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot
ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel ="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of cholesterol, age, thalach(max heart rate)
fig, (ax0,ax2)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True)

# Add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
# Add a meanline
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")

# Add data to ax2
scatter= ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax2
ax2.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax2
ax2.legend(*scatter.legend_elements(),title="Target")
# Add a meanline
ax2.axhline(over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# reset the style
plt.style.use("seaborn-whitegrid")

## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); # this changes the color scheme == cmap is colormap
# Customize the plot
ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel ="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customising the y and x axis limitations


# Subplot of cholesterol, age, thalach(max heart rate)
fig, (ax0,ax2)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True)

# Add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Change the x axis limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
# Add a meanline
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")

# Add data to ax2
scatter= ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                    cmap="winter")

# Customize ax2
ax2.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Change the x axis limits
ax2.set_xlim([50,80])
ax2.set_ylim([60,200])

# Add a legend to ax2
ax2.legend(*scatter.legend_elements(),title="Target")
# Add a meanline
ax2.axhline(over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
